```python
def linear_regression(x, y):     
    N = len(x)
    x_mean = x.mean()
    y_mean = y.mean()
    
    B1_num = ((x - x_mean) * (y - y_mean)).sum()
    B1_den = ((x - x_mean)**2).sum()
    B1 = B1_num / B1_den
    
    B0 = y_mean - (B1*x_mean)
    
    reg_line = 'y = {} + {}β'.format(B0, B1)
    return (B0, B1, reg_line)

x = np.array([25, 50, 75, 100, 150, 200, 250]).transpose() * np.expm1(1e-6)
x = x.reshape((-1, 1))
for n in range(6):
    y = XY_Transpose.iloc[:, [n]].to_numpy() # 0 = Y1   
    C0, C1, reg_line =linear_regression(x, y)
    #print('Regression Line: ', reg_line)
    print('Z = ', Z_Group[n] )            
    print('C1 = ', C1)
    print('C0 = ' , C0)
    print('NDL = ', NDL)
    print('d = ', d)
```
